```python
# ======================================
# IMPORTING REQUIRED MODULES
# ======================================

import os
import requests
from datetime import datetime

# =======================================
# Logger Function ( To help with Task 4)
# =======================================

def log_error(function_name, error_type, message):
    """
    This function logs errors into error_log.txt
    Each log entry includes timestamp, function name, and error message
    """
    with open("error_log.txt", "a", encoding="utf-8") as file:
        time_stamp = datetime.now().strftime("%Y-%m-%d, %H:%M:%S")
        file.write(f"[{time_stamp}] ERROR in {function_name}: {error_type} - {message}\n")


# ========================================
# TASK 1
# ========================================
def file_operations_demo():
    filename ="python_notes.txt"

    # Creating a list of notes
    initial_notes = [
        "Topic 1: Variables store data. Python is dynamically typed.",
        "Topic 2: Lists are ordered and mutable.",
        "Topic 3: Dictionaries store key-value pairs.",
        "Topic 4: Loops automate repetitive tasks.",
        "Topic 5: Exception handling prevents crashes."
    ]
    
    try:
        # Writing to file
        with open(filename, "w", encoding= "utf-8") as f:
            for note in initial_notes:  
                f.write(note + "\n") # Adds new line after each line

            print("File written successfully.")

            # Appending extra lines (adding more notes later)
            extra_notes = [
                "Topic 6: Functions promote code reuse.",
                "Topic 7: APIs allow communmication between systems."
            ]        

        with open(filename, "a", encoding="utf-8") as f:
            for note in extra_notes:
                f.write(note + "\n")

        print("Extra notes appended.")

        # Reading file
        print("\n----- Reading File Contents -----")

        if os.path.exists(filename):
            with open(filename, "r", encoding="utf-8") as f:
                lines = f.readlines()

            # Numbered Output
            for index, line in enumerate(lines, start=1):
                print(f"{index}. {line.strip()}")

            # Count lines
            print(f"\nTotal lines: {len(lines)}")

            # Keyword search 
            keyword = input("\nEnter keyword to search: ").lower()

            found = False
            for line in lines:
                if keyword in line.lower():
                    print(line.strip())
                    found = True
                
            if not found:
                print("No matching lines found.")

    except Exception as e:
        print("Error during file operations: ", e)
        log_error("file_operation_demo", "Exception", str(e))



file_operations_demo()
```

```python
# ========================================
# TASK 2
# ========================================

import requests
def api_operations():

    # Step1: Fetching 20 Products
    try:
        print("\n----- Step 1: Fetching 20 Products -----")

        url = "https://dummyjson.com/products?limit=20"
        response = requests.get(url, timeout=5)
        data = response.json()
        products = data["products"]

        # Table Header
        print("ID   | Title                          | Category        | Price    | Rating")
        print("---------------------------------------------------------------------------")

        for p in products:
            print(f"{p['id']:<4} | {p['title'][:30]:<30} | {p['category']:<14} | ${p['price']:<8} | {p['rating']:<6}")
    
    except requests.exceptions.ConnectionError:
        print("Connection failed. Please Check Your Internet.")
        log_error("fetch_products", "ConnectionError", "No Internet")
        
    except requests.exceptions.Timeout:
        print("Request timed out.")
        log_error("fetch_products", "Timeout", "Slow response")

    except requests.Exception as e:
        print("Unexpected Error.")
        log_error("fetch_products", "Exception", str(e))

    # Step 2: Filtering And Sorting
    try:
        print("\n----- Step2: Filtering And Sorting High Rated products (Rating >=4.5) (In Descending) -----")
        filtered = [p for p in products if p["rating"] >= 4.5]
        filtered.sort(key=lambda x: x["price"], reverse=True)

        for p in filtered:
            print(f"Price: ${p['price']:<8} | Rating: {p['rating']:<5} | Title: {p['title']}")


    except Exception as e:
        print("Error in Filtering: ", e)
        log_error("filter_products", "Eception", str(e))

    # Step 3: Category Searching (Laptop)
    try:
        print("\n----- Step 3: Laptop Category -----")
        
        url = "https://dummyjson.com/products/category/laptops"
        response = requests.get(url, timeout=5)
        data = response.json()

        for p in data["products"]:
            print(f"Laptop: {p['title']:<35} | Price: ${p['price']}")

    except Exception as e:
        print("Error fetching Laptops: ",e)
        log_error("fetch_laptops", "Exception", str(e))

    #  Step 4: POST request 
    try:
        print("\n----- Step 4: Adding A Custom Product(POST) -----")        

        url = "https://dummyjson.com/products/add"

        new_product = {
            "title": "My Custom Product",
            "price": 999,
            "category": "electronics",
            "description": "A product I created via API"
        }

        response = requests.post(url, json=new_product, timeout=5)

        print("\n Response from server: ")
        print(response.json())

    except Exception as e:
        print("POST request failed: ", e)
        log_error("post_product", "Exception", str(e))

api_operations()
```

```python
# ========================================
# TASK 3 - PART A (Guarded Calculator)
# ========================================

def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero."
    except TypeError:
        return "Error: Invalid Input Types."
    
# Testing
print("\n----- Part A: Safe Divide -----")
print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")
print(f"'ten' / 2 = {safe_divide('ten', 2)}")
```

```python
# ========================================
# TASK 3 - PART C & D (API + Input Validation)
# ========================================

import requests
def lookup_products():
    url = "https://dummyjson.com/products"

    print("\n----- Part C & D: Product Lookup -----")

    while True:
        user_input = input("\nEnter Product Id (1-100) or 'quit': ").strip().lower()

        if user_input == "quit":
            print("Exiting lookup tool.")
            break

        # Input Validation
        if not user_input.isdigit():
            print("Warning: Please Enter a Valid Number.")
            continue

        product_id = int(user_input)

        if not (1 <= product_id <= 100):
            print("Waring: ID must be Between 1 and 100.")

        # API CAlling WIth Exception Handling
        try:
            response = requests.get(f"{url}/{product_id}", timeout=5)

            if response.status_code == 404:
                print("Product Not Found.")
            
            elif response.status_code == 200:
                data = response.json()
                print(f"Result: {data['title']} - ${data['price']}")

            else:
                print(f"Uneexpected Status Code: {response.status_code}")
            
        except requests.exceptions.ConnectionError:
            print(f"Connection failed. Please Check your Internet.")

        except requests.exceptions.Timeout:
            print("Request timed out. Try Again Later.")

        except Exception as e:
            print("Unexpected error: ", e)
            
# Calling Function
lookup_products()
```
